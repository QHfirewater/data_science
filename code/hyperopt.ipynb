```python
best = fmin(fn = lambda x :x*x,space = hp.uniform('x',-2.5,2.5),algo = tpe.suggest,max_evals =100)
print(best)
```

```python
def fun_change(x):
    y = (np.sin(x-2))**2 * (np.e) **(-x**2)
    return -y
```

```python
best = fmin(fn = fun_change,space = hp.uniform('x',-2.5,2.5),algo=tpe.suggest,max_evals=100)
```

```python
from sklearn import datasets
from sklearn.preprocessing import normalize,scale
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
import warnings
warnings.filterwarnings('ignore')
from hyperopt import Trials
```

```python
iris = datasets.load_iris()
x = iris.data
y = iris.target
```

```python
def hyperot_model(param):
    x_ = x[:]
    if 'normalize' in param:
        if param['normalize']  == 1:
            x_ = normalize(x_)
        del param['normalize']
    if 'scale' in param:
        if param['scale'] == 1:
            x_ = scale(x)
        del param['scale']
    clf = DecisionTreeClassifier(**param)
    return cross_val_score(clf,x_,y).mean()

space_dtree = {
    'max_depth': hp.choice('max_depth', range(1, 20)),
'max_features': hp.choice('max_features', range(1, 5)),
'criterion': hp.choice('criterion', ["gini", "entropy"]),
'scale': hp.choice('scale', [0, 1]),
'normalize': hp.choice('normalize', [0, 1])
}

def fn_tree(param):
    acc = hyperot_model(param)
    return -acc

```

```python
trials = Trials()
best = fmin(fn=fn_tree,space=space_dtree,algo=tpe.suggest,max_evals=1000,trials = trials)
best
```

```python
params = {'criterion': "gini", 'max_depth': 18, 'max_features': 2}
clf = DecisionTreeClassifier(**params)
print(cross_val_score(clf, x, y).mean())
```

```python
parameters = ['max_depth', 'max_features', 'criterion', 'scale',
              'normalize']  # decision tree
cols = len(parameters)
f, axes = plt.subplots(nrows=1, ncols=cols, figsize=(20, 5))
cmap = plt.cm.jet
for i, val in enumerate(parameters):
    xs = np.array([t['misc']['vals'][val] for t in trials.trials]).ravel()
    ys = [-t['result']['loss'] for t in trials.trials]
    ys = np.array(ys)
    axes[i].scatter(
        xs,
        ys,
        s=20,
        linewidth=0.01,
        alpha=0.5,
        c=cmap(float(i) / len(parameters)))
    axes[i].set_title(val)
```
