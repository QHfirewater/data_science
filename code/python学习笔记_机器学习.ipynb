机器学习的目的是，为那些目前既没有高效的理论支持、又没有高效的计算方法的理论推理与数据探索问题提供解决办法

有监督学习：对数据的若干特征与若干标签之前的关联性进行建模的过程，只要模型被确定，就可以应用到新的未知的数据上,应用的方向主要是分类和回归

分类：可以预测两个或者多个离散分类标签的模型

回归：可以预测连续标签的模型

无监督学习：对不带任何标签的数据特征进行建模，通常被看成是一种让数据自己介绍自己的过程，应用的方向主要是聚类和降维

聚类：监测和识别数据显著组别的模型

降维：从高维数据中检测、识别低维数据结构的模型

特征矩阵，记为X：

样本：即每一行，通常指数据集中的每一个对象

特征：即每一列，通常指每个样本中都具有的某种量化特征观测值

目标数组记为y，目标数组一般是一维数组，其长度等于样本总数，通常采用一维的numpy数组或dataframe的series表示，目标数组通常是我们希望从数据中心预测的量化结果，借用统计学术语，y就是因变量

### sklearn 的评估器API

基本原则：

1️⃣统一性：所有对象使用共同接口连接一组对象方法和统一的文档。

2️⃣内省：所有参数都是公共属性

3️⃣限制对象层级：数据集都用标准的数据类型，参数名称采用标准的python字符串

4️⃣函数组合：许多机器学习任务都可以用一串基本算法实现

5️⃣明智的默认值：当模型需要用户设置参数时，sklearn预先定义适当的默认值


#### API基础知识

常用的步骤如下：

1、通过sklearn 中导入适当的评估器类，选择模型类

2、用合适的数值对模型类进行实例化，配置模型超参数、

3、整理数据，获取特征矩阵和目标数组

4、调用模型的实例的fit（）方法对数据进行拟合

5、对新数据进行应用模型：有监督学习中用predict方法预测新数据标签，无监督学习中用transform和predict转换和推断数据

### 回归：简单线性回归实例

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
rng = np.random.RandomState(42)
```

```python
x = 10*rng.rand(50)
x
```

```python
y = 2*x-1+rng.randn(50)
```

```python
plt.scatter(x,y)#生成需要拟合的数据
```

####  选择模型类

```python
from sklearn.linear_model import LinearRegression
```

#### 将模型的类进行实例化，选择模型超参数

这里的模型类与模型实例不同，根据不同模型的不同情况，需要回答以下问题：

1、是否需要拟合偏移量（即直线的截距）？

2、是否需要对模型进行归一化处理？

3、是否需要对特征进行预处理，以提高模型的灵活性？

4、打算在模型中采用哪些种正则化类型？

5、打算使用多少模型组件？

超参数： 有些参数在选择模型的类的时候就确定好，即在拟合数据之前必须被确定的参数

```python
model = LinearRegression(fit_intercept = True)#确定需要拟合的模型，fit_intercept就是超参数
model
```

#### 将数据整理成特征矩阵和目标数组

```python
x
```

```python
X = x[:,np.newaxis]# 对拟合的数据进行调整
X
```

```python
X.shape
```

```python
y.shape
```

```python
y
```

#### 用模型拟合数据

```python
model.fit(X,y)#fit命令会在模型内部进行大量的运算，运算结果将存储在模型的属性中，供用户使用
```

```python
model.coef_#获取线性拟合的斜率  在sklearn中fit方法获得模型参数都带一条下划线
```

```python
model.intercept_#获取线性拟合的截距
```

#### 预测新数据标签

```python
xfit = np.linspace(-1,11)#确定模拟数据的x值
```

```python
Xfit = xfit[:,np.newaxis]
yfit = model.predict(Xfit)#获得模拟数据相应的y值
```

```python
plt.scatter(x,y)#对原始数据进行绘图
plt.plot(xfit,yfit)#对拟合数据进行绘图
```

### 分类：高斯贝叶斯分类实例

```python
from sklearn.model_selection import train_test_split#导入训练集与测试集分类器
x_iris = iris.drop('species',axis =1)
print(x_iris.shape)
print(iris.shape)
y_iris = iris['species']
print(y_iris.shape)
xtrain,xtest,ytrain,ytest = train_test_split(x_iris,y_iris,random_state =1) #将数据分为训练集和测试集
```

```python
from sklearn.naive_bayes import GaussianNB  #1、选择模型的类
model = GaussianNB()                         #2、初始化模型
model.fit(xtrain,ytrain)                     #3、用模型拟合数据
y_model = model.predict(xtest)              #4、对新数据进行预测
```

```python
#用accuracy_score 工具验证模型预测结果的准确率（预测的结果中，正确的结果占总预测结果样本数的比例）
from sklearn.metrics import accuracy_score 
accuracy_score(ytest,y_model)
```

```python
print(y_model)
print(ytest)
```

### 降维：主成因分析法实例
鸢尾花的数据是思维数据，可以通过降维的方法，在保持数据基本特征的前提下，将数据降低至2维

```python
from sklearn.decomposition import  PCA  #1、选择模型的类  PCA：principal component analysis
model = PCA(n_components=2)              #2、设置超参数，初始化模型
model.fit(x_iris)                        #3、拟合数据，注意这里不要用y变量
x_2d = model.transform(x_iris)           #4、将数据转换为2维数据
```

以下进行验证

```python
print(iris)
iris['PCA1'] = x_2d[:,0]
iris['PCA2'] = x_2d[:,1]
sns.lmplot('PCA1','PCA2',hue ='species',data = iris,fit_reg = False)
#PCA算法根本不知道花的种类标签，但是不同种类的花还是被清晰的区分开来！！！！
```

### 聚类：高斯混合模型的实例

对鸢尾花的种类进行聚类分析


```python
from sklearn.mixture import GaussianMixture                           #1、选择模型的类
model = GaussianMixture(n_components = 3,covariance_type='full')       #2、设置超参数，初始化模型
model.fit(x_iris)                                                      #3、拟合数据，注意不要y变量
y_gmm = model.predict(x_iris)                                          #4、确定簇标签
```

以下为验证操作

```python
y_gmm
```

```python
iris['cluster'] = y_gmm
sns.lmplot('PCA1','PCA2',data = iris,hue = 'species',col = 'cluster',fit_reg = False)
```

## 最优模型的选择
最优模型的问题：找出偏差和方差平衡点的问题

判断一个模型的好坏主要有两大步：

第一步：模型能够很好的描述训练数据的特征，忽略随机偏差

第二步：测试数据能很好适应模型

模型改善的方法：

1、用更复杂/更灵活的模型

2、用更简单/更确定的模型

3、采集更多的训练样本

4、为每个样本采集更多的特征

欠拟合：模型没有足够的灵活性来适应训练数据的所有特征，另一种说法是模型具有高偏差

过拟合：模型过于灵活，在适用训练数据的所有特征的同时，也适应了随机偏差，所以对测试数据的适应程度较小，另一种说法是模型具有高方差

### 特征管道案例

```python
from sklearn.preprocessing import PolynomialFeatures#多项式函数
from sklearn.linear_model import LinearRegression#线性回归方法
from sklearn.pipeline import make_pipeline#加载特征管道
```

```python
def poly(degree = 2):#定义特征管道
    return make_pipeline(PolynomialFeatures(degree),LinearRegression())   #特征管道命令能够将多项式处理和线性拟合封装在一起
```

```python
def make_data(N,err = 1,rseed = 1):#定义生成拟合数据函数
    rng = np.random.RandomState(rseed)
    X = rng.rand(N,1)**2
    y = 10-1/(X.ravel()+0.1)#生成数据
    if err >0:
        y += err *rng.randn(N)#加入误差
    return X,y
```

```python
X,y = make_data(40)#生成需要的数据
print(X)
print(y)
```

### 绘制学习曲线
影响模型效果的因素：模型的复杂程度和训练数据集的规模

学习曲线：反映训练集规模的训练得分和验证得分的曲线

```python
x2,y2 = make_data(200)
plt.scatter(x2.ravel(),y2)
```

学习曲线的特征：

对于特定复杂度的模型对较小的数据集容易过拟合：此时训练得分较高，验证得分较低

特定复杂度的模型对较大的数据集容易欠拟合：此时随着数据的增大，训练得分会不断降低，而验证得分会不断升高

模型的验证集得分永远不会高出训练集得分：两条曲线一直在靠近，但永远不会交叉

```python
from sklearn.model_selection import learning_curve
fig,ax = plt.subplots(1,2,figsize = (16,6))
fig.subplots_adjust(left = 0.0625,right = 0.95,wspace = 0.1)

for i ,degree in enumerate([2,9]):
    N,train_lc,val_lc = learning_curve(poly(degree),X,y,cv= 7,train_sizes = np.linspace(0.3,1,25))
    
    ax[i].plot(N,np.mean(train_lc,1),color = 'blue',label = "train score")
    ax[i].plot(N,np.mean(val_lc,1),color = 'red',label = 'validate score')
    ax[i].hlines(np.mean([train_lc[-1],val_lc[-1]]),N[0],N[-1],color = 'gray',linestyle = 'dashed')
    
    ax[i].set_ylim(0.6,1)
    ax[i].set_xlim(N[0],N[-1])
    ax[i].set_xlabel('train size')
    ax[i].set_ylabel('score')
    ax[i].set_title('degree = {0}'.format(degree),size =14)
    ax[i].legend(loc = 'best')
```

当模型得分已经收敛，则增加训练样本也不会使得分再次升高，这个时候就只能换模型了

当采用更为复杂的模型时，虽然学习曲线的收敛得分提高了，但是模型的方差也变大了（训练得分与收敛结果的差值）

### 高斯朴素贝叶斯分类
贝叶斯定理（条件概率公式）：P(L|特征)= P(特征|L)P(L)/P(特征） 

P(L|特征):在特征发生的条件下，L发生的概率

P(L):在一定的条件下，发生L的概率

P(特征):在一定的条件下，发生特征的概率

朴素贝叶斯的应用场景

缺点：朴素贝叶斯分类器对数据有着严格的假设，因此它的训练效果通常比复杂模型的差

优点：

1、训练和预测的速度非常快

2、直接使用概率预测

3、通常很容易解释

4、可调参数非常少

适用场景：假设分布函数与数据相匹配；各种类型的区分度很高，模型复杂度不重要；非常高维度的数据，模型复杂度不重要

后面两题本质是相同的，随着数据集维度的增加，任何两点都不太可能逐渐靠近（只有每一个维度上都接近才能逐渐靠近），也就是说新维度的增加，高维度数据的簇中心会比低纬度数据的簇中心更加分散，因此随着数据维度的不断增加，像朴素贝叶斯这样的简单分类器的分类效果会比复杂的分类器具有更好的效果

### 高斯基函数回归

```python
from sklearn.base import BaseEstimator,TransformerMixin
```

#### 创建高斯基函数

```python
class GaussianFeatures(BaseEstimator,TransformerMixin):
    #一维输入均匀分布的高斯特征
    def __init__(self,N,width_factor = 2):
        self.N = N
        self.width_factor = width_factor
        
    @staticmethod
    def _gauss_basis(x,y,width,axis = None):
        arg = (x-y)/width
        return np.exp(-0.5*np.sum(arg**2,axis))
    
    def fit(self,X,y =np.nan):
        self.centers_ = np.linspace(X.min(),X.max(),self.N)
        self.width_ = self.width_factor * (self.centers_[1]-self.centers_[0])
        return self
    
    def transform(self,X):
        return self._gauss_basis(X[:,:,np.newaxis],self.centers_,self.width_,axis = 1)
    
```

```python
GaussianFeatures(20)
```

```python
gauss_model = make_pipeline(GaussianFeatures(20),LinearRegression())
```

```python
gauss_model.fit(x[:,np.newaxis],y)
yfit = gauss_model.predict(xfit[:,np.newaxis])
```

```python
plt.scatter(x,y)
plt.plot(xfit,yfit)
plt.xlim(0,10)
```

这个实例主要是为了说明多项式基函数并不是什么魔法，如果对数据的产生过程有某种直觉，那么 就可以自己先定义一些基函数，然后像这样使用他们

正则化的本质是，通过增加训练误差为代价，来减少测试误差的策略

#### 岭回归

```python
#在线性模型中引入基函数可以让模型变的更加灵活，但是也容易造成过拟合的现象
model = make_pipeline(GaussianFeatures(30),LinearRegression())
model.fit(x[:,np.newaxis],y)
```

```python
plt.scatter(x,y)
plt.plot(xfit,model.predict(xfit[:,np.newaxis]))
plt.xlim(0,10)
plt.ylim(-1.5,1.5)
```

```python
#如果将高斯基函数的系数画出来，就可以看到模型过拟合的原因：为了适应随机误差，从而导致基函数的系数偏大
def basis_plot(model,title = None):
    fig,ax = plt.subplots(2,sharex = True)
    model.fit(x[:,np.newaxis],y)
    ax[0].scatter(x,y)
    ax[0].plot(xfit,model.predict(xfit[:,np.newaxis]))
    ax[0].set(xlabel = 'x',ylabel = 'y',ylim = (-1.5,1.5))
    if title:
        ax[0].set_title(title)
    ax[1].plot(model.steps[0][1].centers_,model.steps[1][1].coef_)
    ax[1].set(xlabel = 'basis location',ylabel = 'coefficient',xlim = (0,10))
```

```python
model = make_pipeline(GaussianFeatures(30),LinearRegression())
basis_plot(model)
#当基函数重叠的时候，通常表现出过拟合：相邻基函数的系数相互抵消
#通过对较大的模型参数进行惩罚，从而可以抑制模型剧烈波动，就可以解决这个问题
```

岭回归的处理方法是:对模型系数的平方和进行惩罚，模型拟合的惩罚项为：p = α∑项平方，其中α是自由参数，用来控制惩罚力度

```python
from sklearn.svm import SVC#‘support vector classifier’
model = SVC(kernel = 'linear',C= 1E10)
model.fit(X,y)
```

```python
def plot_svc_decision_fuction(model,ax = None,plot_support = True):
    #画二维SVC的决策函数
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    #创建评估模型的网络
    x = np.linspace(xlim[0],xlim[1],30)
    y = np.linspace(ylim[0],ylim[1],30)
    Y,X = np.meshgrid(y,x)
    xy = np.vstack([X.ravel(),Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    #画决策边界和边界
    ax.contour(X,Y,P,colors ='k',levels = [-1,0,1],alpha = 0.5,linestyles = [':','-',':'])
    #画支持向量
    if plot_support:
        ax.scatter(model.support_vectors_[:,0],
                  model.support_vectors_[:,1],
                  s = 300,linewidth = 1,facecolors = 'none');
        ax.set_xlim(xlim)
        ax.set_ylim(ylim)
```

```python
plt.scatter(X[:,0],X[:,1],c =y,s =50,cmap='autumn')
plot_svc_decision_fuction(model)
```

```python
#用于拟合的关键支持点，被称为支持向量，sklearn里，支持向量的坐标存放在分类器中
model.support_vectors_
```

```python
#分类器能够成功拟合的关键因素，就是这些支持向量的位置——任何在正确分类一侧远离边界线的点都不会影响拟合结果
#从技术角度考虑，是因为这些点不会对拟合模型的损失函数产生任何影响，只要他们没有跨越边界，他们的位置和数量就都无关紧要
def plot_svm(N=10,ax = None):
    X,y = make_blobs(n_samples =N,centers =2,random_state =0,cluster_std=0.6)
    X = X[:N]
    Y = y[:N]
    model = SVC(kernel = 'linear',C = 1E10)
    model.fit(X,y)
    ax = ax or plt.gca()
    ax.scatter(X[:,0],X[:,1],c= y,s=50,cmap = 'autumn')
    ax.set_xlim(-1,4)
    ax.set_ylim(-1,6)
    plot_svc_decision_fuction(model,ax)
```

```python
def visualize(model,X,y,ax =None,cmap = 'rainbow'):
    ax = ax or plt.gca()
    #画出训练数据
    ax.scatter(X[:,0],X[:,1],c=y,s= 30,cmap =cmap,clim = (y.min(),y.max()),zorder = 3)
    ax.axis('tight')
    ax.axis('off')
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    #用评估器拟合数据
    model.fit(X,y)
    xx,yy = np.meshgrid(np.linspace(*xlim,num =200),np.linspace(*ylim,num =200))
    Z = model.predict(np.c_[xx.ravel(),yy.ravel()]).reshape(xx.shape)
    
    #为结果生成彩色图
    n_class = len(np.unique(y))
    contours = ax.contourf(xx,yy,Z,alpha =0.3,levels = np.arange(n_class+1)-0.5,cmap =cmap,zorder =1)
    ax.set(xlim =xlim,ylim = ylim)
    print(Z)
```

```python
def draw_vector(v0,v1,ax =None):
    ax = ax or plt.gca()
    arrow = dict(arrowstyle = '->',
                linewidth = 2,
                 color = 'black',
                shrinkA = 0,shrinkB = 0)
    ax.annotate('',v1,v0,arrowprops = arrow)
```

```python
plt.scatter(X[:,0],X[:,1],alpha =0.2)
for length,vector in zip(pca.explained_variance_,pca.components_):
    v = vector * 3 * np.sqrt(length)
    draw_vector(pca.mean_,pca.mean_ + v)
    plt.axis('equal')
#这些向量表示数据的主轴，箭头的长度表示数据中各个轴的重要程度，更准确的说，他衡量了数据投影到主轴上的方差的大小
```

### k-mean算法：期望最大化

期望最大化主要分为两个步骤，期望步骤和最大化步骤

1、猜测一些簇中心点

2、重复直至收敛：

a、期望步骤：将点分配至离其最近的簇中心点
b、最大化步骤哦：将簇中心点设置为所有坐标点的平均值

期望步骤，不断更新每一个点是属于那一个簇的期望值，

最大化步骤，计算关于簇中心点的拟合函数值最大化对应的坐标

```python
#一个非常简单的k-means算法
from sklearn.metrics import pairwise_distances_argmin
def find_cluster(X,n_clusters,rseed =2):
    #随机选择簇中心点
    rng = np.random.RandomState(rseed)
    i =rng.permutation(X.shape[0])[:n_clusters]
    centers= X[i]
    while True:
        #2a 基于最近的中心点指定标签
        labels = pairwise_distances_argmin(X,centers)
        
        #2b 根据点的平均值找到新的中心
        new_centers = np.array([X[labels == i].mean(0) for i in range(n_clusters)])
        
        #2c 确认收敛
        if np.all(centers == new_centers):
            break
        centers = new_centers
    return centers,labels
```

```python
centers,labels = find_cluster(X,4)
plt.scatter(X[:,0],X[:,1],c = labels,s = 50,cmap = 'viridis')
```

使用期望最大化算法时注意事项：

1、由于随机数的原因，可能不会达到全军最优结果，因此算法最好采用不同的初始值尝试很多遍，在sklearn中通过n_init参数设置执行次数

2、簇数量必须事先定好

```python
from matplotlib.patches import Ellipse
```

```python
def draw_ellipse(position,covariance,ax = None,**kwargs):
    #用给定的未知和协方差画一个椭圆
    ax = ax or plt.gca()
    #将协方差转为主轴
    if covariance.shape == (2,2):
        u,s,vt = np.linalg.svd(covariance)
        angle = np.degrees(np.arctan2(u[1,0],u[0,0]))
        width,height = 2*np.sqrt(s)
    else:
        angle = 0
        width,height = 2*np.sqrt(covariance)
    #画出椭圆
    for nsig in range(1,4):
        ax.add_patch(Ellipse(position,nsig*width,nsig*height,angle,**kwargs))
```

```python
def plot_gmm(gmm,X,label= True,ax=None):
    ax = ax or plt.gca()
    labels = gmm.fit(X).predict(X)
    if label:
        ax.scatter(X[:,0],X[:,1],c = labels,s=40,cmap = 'viridis',zorder=2)
    else:
        ax.scatter(X[:,0],X[:,1],s = 40,zorder=2)
    ax.axis('equal')
    w_factor = 0.2/gmm.weights_.max()
    for pos,covar,w in zip(gmm.means_,gmm.covariances_,gmm.weights_):
        draw_ellipse(pos,covar,alpha = w*w_factor)
```

```python
gmm = GaussianMixture(n_components=4,random_state=42)
plot_gmm(gmm,X)
```

## 核密度估计

```python
def make_data (N,f=0.3,rseed =1):
    rand = np.random.RandomState(rseed)
    x= rand.randn(N)
    x[int(f*N):]+=5
    return x
x = make_data(1000)
```

```python
hist = plt.hist(x,bins = 20,density=True)
```

```python
density,bins,patches =hist
widths = bins[1:]-bins[:-1]
(density * widths).sum()
```

```python
xd =np.linspace(-4,8,2000)
density = sum((abs(xi-xd)<0.5) for xi in x)
plt.fill_between(xd,density,alpha =0.5)
plt.plot(x,np.full_like(x,-0.1),'|k',markeredgewidth =1)
```

```python
from scipy.stats import norm
x_d = np.linspace(-4,8,1000)
density = sum(norm(xi).pdf(x_d) for xi in x)
```

```python
plt.fill_between(x_d,density,alpha =0.5)
plt.plot(x,np.full_like(x,-0.3),'|k',markeredgewidth =1)
```
