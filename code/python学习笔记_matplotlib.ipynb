## 两种绘图接口

### matlab风格接口

```python
plt.figure()  #创建画板

plt.subplot(2,1,1) #创建第一个坐标轴
plt.plot(x,np.sin(x)) #绘制第一个图形

plt.subplot(2,1,2) #创建第二个坐标轴
plt.plot(x,np.cos(x))#绘制第二个图形
```

这种接口方式会持续跟踪当前的图形和坐标轴

### 面向对象的接口

面向对象的接口方式可以适应更加复杂的场景，可以更好的控制自己的图形

```python
fig,ax = plt.subplots(2) #同时创建画板和两个坐标轴
ax[0].plot(x,np.sin(x)) #在第一个坐标轴中绘图
ax[1].plot(x,np.cos(x)) #在第二个坐标轴中绘图
```

### 选择图例显示的元素

```python
y = np.sin(x[:,np.newaxis]+np.pi*np.arange(0,2,0.5))
lines = plt.plot(x,y)
plt.legend(lines[:2],['first','second']) #自定义显示图例
```

```python
plt.plot(x,y[:,0],label ='first')#另外一种显示不同图例的方法
plt.plot(x,y[:,1],label = 'second')
plt.plot(x,y[:,2:])
plt.legend(framealpha =1,frameon = True)
```

### 设置显示多个图例

```python
fig,ax = plt.subplots()

line =[]
styles = ['-','--','-.',':']
x = np.linspace(0,10,1000)

for i in range(4):
    lines += ax.plot(x,np.sin(x-i*np.pi/2),styles[i],color = 'black')
    
ax.axis('equal')

# 设置第一个图例要显示的线条和标签
ax.legend(lines[:2],['linA','linB'],loc ='upper right',frameon = False)

#创建第二个图例，通过add_artist 方法添加到图上
from matplotlib.legend import Legend
leg = Legend(ax,lines[:2],['linC','linD'],loc = 'lower right',frameon = False)

ax.add_artist(leg)
```

### plt.gridspec 可以实现更加复杂的排列方式

```python
grid = plt.GridSpec(2,3,wspace =0.4,hspace =0.3)
plt.subplot(grid[0,0]) #在第一个位置上绘图
plt.subplot(grid[0,1:]) #在第二和第三个位置是哪个绘图
plt.subplot(grid[1,:2])
plt.subplot(grid[1,2])
```

```python
mean = [0,0]
cov = [[1,1],[1,2]]
x,y = np.random.multivariate_normal(mean,cov,3000).T

fig = plt.figure(figsize = (6,6))
grid = plt.GridSpec(4,4,hspace = 0.2,wspace = 0.2)
main_ax = fig.add_subplot(grid[:-1,1:])
y_hist = fig.add_subplot(grid[:-1,0],xticklabels = [],sharey = main_ax)
x_hist = fig.add_subplot(grid[-1,1:],yticklabels = [],sharex = main_ax)

main_ax.plot(x,y,'ok',markersize = 3,alpha = 0.2)
x_hist.invert_yaxis()
x_hist.hist(x,40,histtype = 'stepfilled',orientation = 'vertical')
y_hist.hist(y,40,histtype = 'stepfilled',orientation = 'horizontal')
y_hist.invert_xaxis()
```

## 文字与注释

### 坐标变换与文字位置

```python
fig,ax = plt.subplots(facecolor ='lightgray')
ax.axis([0,5,0,10])
ax.text(1,5,'.  data:(1,5)',transform = ax.transData) #用x轴和y轴标签作为数据坐标
ax.text(0.5,0.1,'. AXES:(0.5,0.1)',transform =ax.transAxes) #以坐标轴左下角的位置为原点，按坐标轴尺寸比例呈现坐标
ax.text(0.2,0.2,'. figure:(0.2,0.2)',transform = fig.transFigure) #以图形（灰色矩形）左下角的位置为原点，按图形吃尺寸的比例呈现坐标
```

```python
#当改变坐标轴上下限是，只有transData的坐标会受到影响，其他坐标位置不变
ax.set_xlim(0,2)
ax.set_ylim(-6,6)
fig
```

```python
def for_fuc(value,tick_number):
    #找到pi/2的倍数刻度线位置
    N = int(np.round(2 * value/ np.pi))
    if N == 0:
        return '0'
    elif N == 1:
        return r'$\pi/2$'
    elif N == 2:
        return r'$\pi$'
    elif N %2 >0:
        return r'${0}\pi/2$'.format(N)
    else:
        return r'${0}\pi$'.format(N//2)
```

```python
fig,ax = plt.subplots() #设置花哨的刻度线格式
x= np.linspace(0,3*np.pi,1000)
ax.plot(x,np.sin(x),lw = 3,label = 'sin(x)')
ax.plot(x,np.cos(x),lw = 3,label = 'cos(x)')
ax.grid(True)
ax.legend(frameon =False)
ax.axis('equal')
ax.set_xlim(0,3*np.pi)
#为了更好的显示，在pi/4的倍数上添加主要刻度和次要刻度
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi/2))
ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi/4))
# 最优的显示方法
ax.xaxis.set_major_formatter(plt.FuncFormatter(for_fuc))
```

### 绘制莫比乌斯环

```python
theta = np.linspace(0,2*np.pi,30)
w = np.linspace(-0.25,0.25,8)
w,theta = np.meshgrid(w,theta)
```

```python
phi = 0.5 *theta
```

```python
r = 1+ w *np.cos(phi)

x = np.ravel(r *np.cos(theta))
y = np.ravel(r*np.sin(theta))
z = np.ravel(w *np.sin(phi))
```

```python
from matplotlib.tri import Triangulation
```

```python
tri = Triangulation(np.ravel(w),np.ravel(theta))
```

```python
ax = plt.axes(projection = '3d')
ax.plot_trisurf(x,y,z,triangles = tri.triangles,cmap = 'viridis',linewidths = 0.2)
```
